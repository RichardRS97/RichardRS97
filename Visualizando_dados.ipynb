```python
import openpyxl
import pandas as pd





# Abrindo arquivos excel xlsx
df = pd.read_excel(r"C:\Users\Richard\OneDrive\Área de Trabalho\ESTUDOS\df.xlsx")

df.sample(5)
```

```python
# Value Counts
df['LojaID'].value_counts(ascending=False) # ascending=False para ordenar do maior para o menor

# Assim é observado que a Loja 1036 realizou 117 vendas
```

```python
# Importando a biblioteca matplotlib
!pip install matplotlib

import matplotlib.pyplot 
# Criando um gráfico de barras Vendas x LojaID
df['LojaID'].value_counts().plot.bar()
```

```python
# Criando um gráfico de barras horizontais
df['LojaID'].value_counts(ascending=True).plot.barh()
```

```python
# Gráfico de Pizza
df.groupby(df['Data'].dt.year)['Receita'].sum().plot.pie() # Agrupando por ano e somando a receita
```

```python
# Total de vendas por cidade
df['Cidade'].value_counts()
```

```python
# Adicionando um título e alterando o nome dos eixos
import matplotlib .pyplot as plt
df['Cidade'].value_counts().plot.bar(title="Total Vendas por Cidade")
plt.xlabel('Cidade') # Alterando o nome do eixo x
plt.ylabel('Total Vendas'); # Alterando o nome do eixo y
```

```python
# Alterando a cor 
df['Cidade'].value_counts().plot.bar(title="Total Vendas por Cidade", color='red')
```

```python
# Alterando o estilo
plt.style.use('ggplot')
```

```python
# Novo gráfico
df.groupby(df["mes_venda"])["Qtde"].sum().plot(title = "Total Produtos Vendidos x Mês")
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos")
plt.legend(); # Adicionando legenda
```

```python
# Selecionando apenas as vendas de 2019
df_2019 = df[df["Ano_Venda"] == 2019]
```

```python
# Total de produtos vendidos por mês
df_2019.groupby(df_2019["mes_venda"])["Qtde"].sum().plot(marker = "v")
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos")
plt.legend()
```

```python
# histograma 
plt.hist(df["Qtde"], color="orange", edgecolor="black")
```

```python
# Dispersão
plt.scatter(x=df_2019["dia_venda"], y = df_2019["Receita"]);
```

```python
# Salvando em png
import matplotlib.pyplot as plt
df_2019.groupby(df_2019["mes_venda"])["Qtde"].sum().plot(marker = "v")
plt.title("Quantidade de vendas x mês")
plt.xlabel("Mês")
plt.ylabel("Quantidade")
plt.savefig("vendas_por_mes.png", format = "png")
```
